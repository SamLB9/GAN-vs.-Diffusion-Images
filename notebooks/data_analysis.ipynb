```python
import os
import random

# Path to your ImageNet class folders
base_dir = "/Users/samlaborde-balen/Desktop/IPP/src/data/BEST/everything/real_imagenet"

# Loop through class folders: '00000' to '00999'
for class_id in range(1000):
    folder_name = f"{class_id:05d}"  # zero-padded 5-digit format
    class_folder = os.path.join(base_dir, folder_name)

    if not os.path.isdir(class_folder):
        print(f"Skipping missing folder: {class_folder}")
        continue

    # Get image files (adjust extensions if needed)
    image_files = [f for f in os.listdir(class_folder) if f.lower().endswith(('.jpg', '.jpeg', '.png'))]

    # Keep only 10 images
    if len(image_files) > 10:
        keep_files = set(random.sample(image_files, 10))
        for img in image_files:
            if img not in keep_files:
                os.remove(os.path.join(class_folder, img))
        print(f"✅ {folder_name}: kept 10, deleted {len(image_files) - 10}")
    else:
        print(f"ℹ️  {folder_name}: already has ≤10 images")

print("✅ Done trimming all class folders to 10 images each.")
```

```python
#!/usr/bin/env python3
"""
Batch-resize all images in a folder to 200×200 pixels.

Usage:
    1. Install Pillow if you haven’t already:
         pip install pillow
    2. Edit INPUT_DIR to point at your source images.
    3. (Optional) Edit OUTPUT_DIR if you don’t want to overwrite originals.
    4. Run:
         python resize_images.py
"""

import os
from PIL import Image

# ----- CONFIGURE THESE -----
INPUT_DIR = './data/BEST/face/real'      # folder containing your images
OUTPUT_DIR = INPUT_DIR       # set to a different path to keep originals
TARGET_SIZE = (200, 200)     # (width, height)
# ---------------------------

# create output directory if needed
if OUTPUT_DIR != INPUT_DIR and not os.path.exists(OUTPUT_DIR):
    os.makedirs(OUTPUT_DIR)

# supported extensions
EXTENSIONS = ('.jpg', '.jpeg', '.png', '.bmp', '.tiff', '.gif')

for fname in os.listdir(INPUT_DIR):
    if not fname.lower().endswith(EXTENSIONS):
        continue

    in_path = os.path.join(INPUT_DIR, fname)
    out_path = os.path.join(OUTPUT_DIR, fname)

    try:
        with Image.open(in_path) as img:
            # Resize with Lanczos filter for high quality
            img_resized = img.resize(TARGET_SIZE, Image.LANCZOS)
            # Preserve original format
            img_resized.save(out_path, img.format)
            print(f"Resized {fname} → {TARGET_SIZE[0]}×{TARGET_SIZE[1]}")
    except Exception as e:
        print(f"Failed to process {fname}: {e}")
```

```python
#!/usr/bin/env python3
"""
Recursively resize and pad all images in a directory tree to 256×256,
preserving aspect ratio and subfolder structure.

Usage:
    1. pip install pillow
    2. Edit INPUT_ROOT and OUTPUT_ROOT.
    3. Run:
         python resize_and_pad.py
"""

import os
from PIL import Image

# ----- CONFIGURE THESE -----
INPUT_ROOT  = 'data/BEST/everything/real_imagenet/'        # root containing your 1000 subfolders
OUTPUT_ROOT = 'data/BEST/everything/real_imagenet/' # where resized trees will go
TARGET_SIZE = (200, 200)                # (width, height)
PADDING_COLOR = (0, 0, 0)               # RGB tuple for pad (black); change as needed
# ---------------------------

# supported image file extensions
EXTENSIONS = ('.jpg', '.jpeg', '.png', '.bmp', '.tiff', '.gif')

for dirpath, _, filenames in os.walk(INPUT_ROOT):
    # compute relative folder path and corresponding output folder
    rel_dir = os.path.relpath(dirpath, INPUT_ROOT)
    out_dir = os.path.join(OUTPUT_ROOT, rel_dir)
    os.makedirs(out_dir, exist_ok=True)

    for fname in filenames:
        if not fname.lower().endswith(EXTENSIONS):
            continue

        in_path  = os.path.join(dirpath, fname)
        out_path = os.path.join(out_dir, fname)

        try:
            with Image.open(in_path) as img:
                img = img.convert('RGB')
                # Compute the new size, preserving aspect ratio
                img.thumbnail(TARGET_SIZE, Image.LANCZOS)

                # Create a new image and paste the thumbnail centered
                background = Image.new('RGB', TARGET_SIZE, PADDING_COLOR)
                x = (TARGET_SIZE[0] - img.width) // 2
                y = (TARGET_SIZE[1] - img.height) // 2
                background.paste(img, (x, y))

                # Save with same format
                background.save(out_path, format=img.format)
                print(f"Processed: {os.path.join(rel_dir, fname)} → {TARGET_SIZE[0]}×{TARGET_SIZE[1]}")
        except Exception as e:
            print(f"Error processing {in_path}: {e}")
```
